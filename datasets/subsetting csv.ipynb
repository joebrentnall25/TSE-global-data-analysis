```python
import pandas as pd

df = pd.read_csv('./proper.csv', index_col=0)
df
```

```python
# remove any cols from the dataset where the mean number of NaN is greater than 25%
df = df[df.columns[df.isnull().mean() < 0.25]]
```

```python
# Subset by Yearcode
dfOverall = df.loc[(df['Time Code'] >= 'YR2007') & (df['Time Code'] <= 'YR2017')]

df2017 = df.loc[df['Time Code'] == 'YR2017']
df2016 = df.loc[df['Time Code'] == 'YR2016']
df2015 = df.loc[df['Time Code'] == 'YR2015']
df2014 = df.loc[df['Time Code'] == 'YR2014']
df2013 = df.loc[df['Time Code'] == 'YR2013']
df2012 = df.loc[df['Time Code'] == 'YR2012']
df2011 = df.loc[df['Time Code'] == 'YR2011']
df2010 = df.loc[df['Time Code'] == 'YR2010']
df2009 = df.loc[df['Time Code'] == 'YR2009']
df2008 = df.loc[df['Time Code'] == 'YR2008']
df2007 = df.loc[df['Time Code'] == 'YR2007']

dfOverall
```

```python
# Drop the time code field from the induvidual year Dataframes, but keep it in overall
df2017.drop(['Time Code'], axis=1)
df2016.drop(['Time Code'], axis=1)
df2015.drop(['Time Code'], axis=1)
df2014.drop(['Time Code'], axis=1)
df2013.drop(['Time Code'], axis=1)
df2012.drop(['Time Code'], axis=1)
df2011.drop(['Time Code'], axis=1)
df2010.drop(['Time Code'], axis=1)
df2009.drop(['Time Code'], axis=1)
df2008.drop(['Time Code'], axis=1)
df2007.drop(['Time Code'], axis=1)
```

```python
# write out new datasets
df2017.to_csv('./worldBank2017.csv', index=False)
df2016.to_csv('./worldBank2016.csv', index=False)
df2015.to_csv('./worldBank2015.csv', index=False)
df2014.to_csv('./worldBank2014.csv', index=False)
df2013.to_csv('./worldBank2013.csv', index=False)
df2012.to_csv('./worldBank2012.csv', index=False)
df2011.to_csv('./worldBank2011.csv', index=False)
df2010.to_csv('./worldBank2010.csv', index=False)
df2009.to_csv('./worldBank2009.csv', index=False)
df2008.to_csv('./worldBank2008.csv', index=False)
df2007.to_csv('./worldBank2007.csv', index=False)
```
